```python
import json
import urllib.request
import requests
```

```python
Url = "http://fxsim.info:3000/user/maruyama/case/test"
def get_data():
    res = urllib.request.urlopen(Url)
    return json.load(res)

def submit_data(ret):
    s = requests.session()
    r =  s.post(Url, data=json.dumps(ret))
    print(r.text.encode("utf-8"))
```

```python
# input as a nested map
input_data_map = get_data()
```

```python
count = 0

tx_list = []
profit = []
position = {"USD": 1000}
for index, row  in data.iterrows():
    tx_receipt = {}
    if count == 0:
        before_ser = row
        before_idx = index
    else:
        diff = -1*(row-before_ser)/before_ser
#         print(index)
        tx_receipt["date"] = before_idx
        
        buy_curr = diff.idxmax()
        position_currency = list(position.keys())[0]
#         print(position)
        
        if buy_curr != position_currency:
            
            position_currency_amount = position.pop(list(position.keys())[0])
            trade = [{"buy": buy_curr,
                      "sellamount": position_currency_amount,
                      "sell": position_currency}]


            if position_currency == "USD":
                position[buy_curr] = (before_ser[buy_curr]*position_currency_amount).round(4)
                tx_receipt["trades"] = trade
                tx_receipt["balance"] = {buy_curr: position[buy_curr]}
            else:
                usd = (position_currency_amount/before_ser[position_currency]).round(4)
                position[buy_curr] = (before_ser[buy_curr]*usd).round(4)
                buy_pos = (before_ser[buy_curr]*usd).round(4)
                tx_receipt["trades"] = trade
                next_position = {buy_curr: buy_pos}
                pl = []
                pl.append(next_position)
                tx_receipt["balance"] = next_position
                profit.append(usd)
                next_position = {}
                pl = []
                if count == len(data)-1:
                    print(usd)

            tx_list.append(tx_receipt)
        
        
        before_ser = row
        before_idx = index
#         print(tx_list)
    

        
    count = count+1
```
