```python
import json
import urllib.request
import requests
```

```python
Url = "http://fxsim.info:3000/user/<userid>/case/test"
def get_data():
    res = urllib.request.urlopen(Url)
    return json.load(res)

def submit_data(ret):
    s = requests.session()
    r =  s.post(Url, data=json.dumps(ret))
    print(r.text.encode("utf-8"))
```

```python
# input as a nested map
input_data_map = get_data()
```
