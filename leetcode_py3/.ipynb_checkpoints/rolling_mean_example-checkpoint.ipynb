```python
import pandas as pd
import requests
import datetime
import matplotlib.pyplot as plt
```

```python
mkt_data = requests.get("http://fxsim.info:5000/data").json()
print(mkt_data)
```

```python
df = pd.DataFrame(mkt_data, dtype=float).T
df.index = pd.to_datetime(df.index, format='%Y%m%d')
```

```python
df.head()
```

```python
# jpy in 2014
first_date = datetime.datetime(2014,1,1)
last_date = datetime.datetime(2014,12,31)
jpy = df["JPY"][(df.index >= first_date) & (df.index <= last_date)]
```

```python
jpy.plot()
for window in ["5d", "25d", "75d"]:
    jpy.rolling(window).mean().plot(label=window)
plt.legend()
```
